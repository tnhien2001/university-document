# Read data and look statistics

```python
import pandas as pd
data = pd.read_csv("Telco Customer Churn.csv")

print("Display all first of 5 rows :")
display(data.head())
print("The shape of data in (nrows,ncols)")
print(data.shape)
```

# Declare required libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# Show data 

```python
print("How to extract Index of Dataframe ? \n\t", data.index)
print("\nHow to extract Column of Dataframe in the list type ? \n\t", list(data.columns))
print(f"\nWhat is min max value of MonthlyCharges in the data ? \n\t  \
      from min : {np.min(data['MonthlyCharges'])} to max : {np.max(data['MonthlyCharges'])}")
print(f"\nWhat is sum and median value of tenure in the data ? \n\t  \
      sum : {data['tenure'].sum()} median : {data['tenure'].median()}")
print(f"\nWhat is sum and median value of TotalCharges in the data ? \n\t  \
      sum : {data['MonthlyCharges'].sum()} median : {data['MonthlyCharges'].median()} std : {data['MonthlyCharges'].std()}")
```

```python
print("How many cases of Churn ?")
print(data["Churn"].value_counts())
print("\n")

print("Can we see the statistics table of the whole data ?")
display(data.describe())
print("\n")

print("Is there any missing value at all columns ?")
display(data.isnull().sum())
print("\n")
```

```python
print("How many customer use PhoneService in data")
print(data["PhoneService"].value_counts())
print("\n")

print("How we can get only the rows from index 100 to 105 ?")
display(data.loc[100:105, :])
print("\n")

print("How we can reset index of the above results in a new table ? ")
df = data.loc[10:15, :]
df = df.reset_index(drop = True)
display(df)
print("\n")

print("How we can get the rows from index 100 to 105 of columns customerID, MonthlyCharges, TotalCharges and Churn ?")
display(data.loc[10:15, ["customerID", "MonthlyCharges", "TotalCharges", "Churn"]])
print("\n")

print("How we can get only the rows from index 110 of columns customerID, MonthlyCharges, TotalCharges and Churn ?")
display(data.loc[110, ["customerID", "MonthlyCharges", "TotalCharges", "Churn"]])
```

```python
print("How we can see the type of all columns in data ?")
display(data.info())
print("\n")

print("How we can change TotalCharges from object to float64")
print("Original Type of TotalCharges :", data["TotalCharges"].dtypes)
data["TotalCharges"] = pd.to_numeric(data['TotalCharges'],errors='coerce')
print("New Type of TotalCharges :", data["TotalCharges"].dtypes)
print("\n")

print("Is there any missing value at all columns ?")
display(data.isnull().sum())
print("\n")

print("How we can extract the categorical and numeric columns ?")
CatFeatures = [col for col in data.columns if data[col].dtypes in ["object", "bool"]]
NumFeatures = [col for col in data.columns if data[col].dtypes in ["int64", "float64"]]
print("Categorical Features :\n", CatFeatures)
print("Numeric Features :\n", NumFeatures)
print("\n")

print("How we can show the all statistics of Numeric Features ?")
display(data.describe())
print("\n")

print("How we can show the all statistics of Categorical Features ?")
display(data[CatFeatures].describe(include='all'))
print("\n")

print("How we can get data from describe table ?")
NumStats = data[NumFeatures].describe(include='all')
CatStats = data[CatFeatures].describe(include='all')
min_max_TotalCharges = NumStats.loc[["min","max"],"TotalCharges"]
Contract_top_freq = CatStats.loc[["top", "freq"], "Contract"]
print("Min and Max of TotalCharges: \n", min_max_TotalCharges)
print("\n")
print("Top and Frequency of Contract: \n",Contract_top_freq)
```

# Fix missing data

```python
#Find the index of missing data
print("The index of missing data are: ",np.where(data['TotalCharges'].isnull())[0])
print("The data of the index 488 is: ",data['TotalCharges'][488])
print("The index 487: ",data['TotalCharges'][487], "\nThe index 489: ",data['TotalCharges'][489])
new_data = data

#Fix the missing data with the mean of between data
new_data = new_data.interpolate()
print("The data of the index 488 is: ",new_data['TotalCharges'][488])
print("Is there any missing value at all columns in new_data ?")
display(new_data.isnull().sum())
print("\n")


#Assign new data to old data
data = new_data
```

```python
print("How we can draw chart for a numeric features ?")
feature = "MonthlyCharges"
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
f.set_figheight(3)
f.set_figwidth(15)
sns.boxplot(data[feature], ax=ax_box)
sns.histplot(data=data, x=feature, ax=ax_hist)
plt.grid()
plt.show()

print("How we can map Yes/No to True/False in Churn feature  ?")
MapDict = {"Yes" : True, "No": False}
data["Churn_Or_Not"] = data["Churn"].map(MapDict)
display(data.head())

print("How we can draw chart for a numeric feature according to a categorical feature ?")
feature = "MonthlyCharges"
plt.figure(figsize = (15,3))
sns.boxplot(y ='Churn', x = feature, data = data)
plt.title(feature)
plt.grid()
plt.show()

print("How we can draw chart for two numeric features according to a categorical feature ?")
plt.figure(figsize=(15,5))
feature_x = "tenure"
feature_y = "MonthlyCharges"
feature_hue = "Churn"
sns.scatterplot(x = feature_x, y= feature_y, hue=feature_hue, data = data, legend='full')
plt.grid()
plt.show()
```

```python
print("How we can draw chart for a categorical feature ?")
feature = "Churn"
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
data[feature].value_counts().plot.pie(autopct='%1.1f%%')
plt.subplot(1,2,2)
sns.countplot(data[feature])
plt.show()

print("How we can draw chart for a categorical feature according to another categorical feature ?")
plt.figure(figsize=(15,5))
feature_x = "PhoneService"
feature_y = "MonthlyCharges"
sns.stripplot(data[feature_x],data[feature_y])
plt.show()
```

```python
print("How we can split data into many data ? ")
feature_Customer = ['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure','Churn']
data_Customer = data[feature_Customer].copy()
display(data_Customer.head())

feature_Service = ['PhoneService', 'MultipleLines', 'InternetService', "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"]
data_Service = data[feature_Service].copy()
display(data_Service.head())

feature_Fee = list(set(data.columns) - set(feature_Customer) - set(feature_Service))
data_Fee = data[feature_Fee].copy()
display(data_Fee.head())

print("How we can merge two data into one by cols ? ")
data_Customer_Service = pd.concat([data_Customer, data_Service], axis = 1)
display(data_Customer_Service.head())
```

```python
print("How we can filter data by condition ?")
print("Filter the data of SeniorCitizen > 0 and PhoneService is 'Yes'")
Condition1 = data["SeniorCitizen"] > 0
Condition2 = data["PhoneService"] == "Yes"
data_over100_PhoneService1 = data[Condition1 & Condition2].copy()
display(data_over100_PhoneService1.head())
print(data_over100_PhoneService1.shape)

print("Filter the data of SeniorCitizen < 0 and PhoneService is 'Yes'")
value1, value2 = 1 , "Yes"
data_less1_PhoneService2 = data.query("`SeniorCitizen` < @value1 and `PhoneService` == @value2")
display(data_less1_PhoneService2.head())
print(data_less1_PhoneService2.shape)

print("How we can merge two data into one by rows ?")
data_merge = pd.concat([data_over100_PhoneService1, data_less1_PhoneService2])
display(data_merge)
print(data_merge.shape)
```

# Group by DataFrame

```python
print("How we can group by customerID and Max on tenure, Sum on SeniorCitizen and Average on TotalCharges")
customerDF = pd.DataFrame()
customerDF['Max on tenure'] = data.groupby('customerID')['tenure'].max()
customerDF['Sum on Senior Citizen'] = data.groupby('customerID')['SeniorCitizen'].sum()
customerDF['Average on Monthly Charges'] = data.groupby('customerID')['MonthlyCharges'].mean()
customerDF['Meadian on Total Charges'] = data.groupby('customerID')['TotalCharges'].median()
display(customerDF)

print("How we can join data with customerDF on the customer information to create new information about customer")
customerDF["customerID"] = customerDF.index
customerDF = customerDF.reset_index(drop = True)
display(customerDF.head())
data = pd.merge(data, customerDF, left_on='customerID', right_on='customerID')
display(data.head())
```
