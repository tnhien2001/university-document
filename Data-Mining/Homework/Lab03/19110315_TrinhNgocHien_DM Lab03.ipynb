# Question 1: Visualization the column x in Anscombe Dataset 

```python
#Axis x
nbins = 15
arr = np.array(anscombe['x'])
bins = np.linspace(arr.min() - 1, arr.max(), nbins + 1)
print(f"Number of interval {nbins} bins is calculated : ")
display(bins)
anscombe['binned'] = pd.cut(arr, bins)
display(anscombe.head())
print(anscombe["binned"].value_counts())
```

```python
bins
```

```python
figure(figsize=(20, 5))
plt.title("Histogram of X", size = 15)
plt.xlabel("Bin of X", size = 15)
plt.ylabel("Frequency of each bins", size = 15)
density, bins, _= plt.hist(anscombe['x'], bins = 15, alpha=0.5, color = "b")
for x,y in zip(bins, density):
    if y != 0:
        plt.text(x+0.3, y+0.05, y, fontsize=15) # x,y,str 
plt.show()
```

```python
figure(figsize=(20, 5))
plt.title("Count Plot of X by Bins", size = 15)
ax = sns.countplot(x="binned", data=anscombe, alpha=0.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.xlabel("Bin of X", size = 15)
plt.ylabel("Count of each bins", size = 15)

for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(anscombe))
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')

plt.show()
```

# Question 2: Visualization the column Pulse in Exercise Dataset 

## Comparision of Pulse of Low Fat and No Fat

```python
exercise_low_fat = exercise[exercise['diet'] == "low fat"]
display(exercise_low_fat.head())
exercise_no_fat = exercise[exercise['diet'] == "no fat"]
display(exercise_no_fat.head())
```

```python
fig, axes  = plt.subplots(nrows=2, ncols=2)
fig.set_figwidth(12)
fig.set_figheight(10)
fig.suptitle("Bi-Variate Analysis")

sns.histplot(exercise_low_fat['pulse'], ax=axes[0, 0], kde = True, color = 'r')
sns.histplot(exercise_no_fat['pulse'], ax=axes[0, 0], kde = True , color = 'b'); 
axes[0, 0].set_title('Low Fat and No Fat Diet')

ax_rowspan = plt.subplot2grid((2, 2), (0, 1), rowspan=2)
sns.boxplot(x="time", y="pulse", hue="kind", data=exercise, palette="Set2", ax=ax_rowspan)


sns.scatterplot(data=exercise, x="id", y="pulse", hue="kind", ax=axes[1, 0])
axes[1, 0].set_title('Distribution ID and Pulse')

plt.show()
```

```python
fig, axes  = plt.subplots(nrows=2, ncols=2)
fig.set_figwidth(20)
fig.set_figheight(15)
fig.suptitle("Bi-Variate Analysis")

ax_colspan = plt.subplot2grid((2, 2), (0, 0), colspan=2)
sns.histplot(exercise_low_fat['pulse'], kde = True, color = 'r', ax= ax_colspan)
sns.histplot(exercise_no_fat['pulse'], kde = True , color = 'b', ax= ax_colspan); 
ax_colspan.set_title('Low Fat and No Fat Diet')

sns.boxplot(x="time", y="pulse", hue="kind", data=exercise, palette="Set3", ax=axes[1, 1])
axes[1, 1].set_title('Box plot of Time and Pulse')

sns.scatterplot(data=exercise, x="id", y="pulse", hue="kind", ax=axes[1, 0])
axes[1, 0].set_title('Distribution ID and Pulse')

plt.show()
```

## Assess
- The pulse in low fat is less than the pulse in no fat.
- The pulse in low fat reaches its minimum value at 80 and reaches its maximum value at 132.
- The pulse in no fat reaches its minimum value at 83 and reaches its maximum value at 150.
- Surveys show that the number of ID running reach pulse levels in the range of [80, 150].
- Surveys show that the number of ID resting reach pulse levels in the range of [80, 100].
- Surveys show that the number of ID walking reach pulse levels in the range of [80, 110).

```python
attention_wide = attention.pivot("subject", "solutions", "score")
attention_wide.head()
```

```python
import pandas as pd

# Setup the subplot2grid Layout
fig = plt.figure(figsize=(15, 12))
ax1 = plt.subplot2grid((3,5), (0,0), colspan = 3, rowspan = 2) 
ax2 = plt.subplot2grid((3,5), (0,3), colspan = 2, rowspan = 2)
ax3 = plt.subplot2grid((3,5), (2,0), colspan = 2)
ax4 = plt.subplot2grid((3,5), (2,2), colspan = 3)

# Time Series
sns.lineplot(data=attention, x="subject", y="score", hue="solutions", ax = ax1)
sns.scatterplot(data=attention, x="subject", y="score", hue="solutions",legend=False, ax = ax1)
ax1.set_title("Time Series")

# Scatterplot
sns.heatmap(attention_wide, annot=True, ax = ax2)
ax2.set_title("Heatmap")

# BarPlot
sns.barplot(x='subject',y='score',data=attention, ax = ax3)
ax3.set_title("Barplot")

# Correlation
sns.scatterplot(data=attention, x="subject", y="solutions", size="score", legend=False, sizes=(20, 2000), ax = ax4)

fig.tight_layout()
```

```python
group_by_exercise = exercise.groupby('kind').mean()
group_by_exercise[''] = group_by_exercise[group_by_exercise.columns[0]] #duplicating the first column

angle = np.linspace(start=0, stop=2 * np.pi, num=len(group_by_exercise.columns))

plt.figure(figsize=(5, 5))
plt.subplot(polar=True)

values = group_by_exercise[:1].values[0]
plt.plot(angle, values, label='Rest', color='r')
plt.fill(angle, values, 'r', alpha=0.2)

values = group_by_exercise[1:2].values[0]
plt.plot(angle, values, label='Walking', color='g')
plt.fill(angle, values, 'g', alpha=0.2)

values = group_by_exercise[2:3].values[0]
plt.plot(angle, values, label='Running', color='b')
plt.fill(angle, values, 'b', alpha=0.2)

#labels
plt.title('Kind comparison', size=15, loc='left')
labels = plt.thetagrids(np.degrees(angle), labels=group_by_exercise.columns.values)
plt.legend()

plt.show()
```
