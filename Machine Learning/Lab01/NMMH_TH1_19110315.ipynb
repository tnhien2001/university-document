```python
import pandas as pd
df = pd.read_csv("House_Price_2.csv") 
df.head()
```

```python
x1 = df['width'].values  
print(x1.shape)  ## trích width thành một mảng numpy
x1 = x1.reshape(-1, 1)                     ## chuyển thành ma trận cột
print(x1.shape)
```

```python
x2 = df['length'].values   ## Trích length 
print(x2.shape)
x2 = x2.reshape(-1,1)
print(x2.shape)
```

```python
y = df['price'].values   ## Trích giá tiền 
print(y.shape)
y = y.reshape(-1,1)
print(y.shape)
```

```python
import numpy as np
Z = np.concatenate([x1,x2, np.ones((100,1))], axis=-1)   ## thêm một cột vào sau X và gán vào Z
Z = Z.astype(np.float32)                                 ## chuyển sang float
print(Z)
```

```python
#Từ đó ta tính y = (a1x1 + a2x2) + b
#Giá trị a1, a2  và b tìm được
a_1, a_2, b = np.matmul(np.matmul(np.linalg.inv(np.matmul(Z.T,Z)), Z.T), y)
print(a_1, a_2, b)
```

```python
#Trường hợp width = 5 length = 10

w_1 = 5
l_1 = 10
price_1 = (a_1 * w_1 + a_2 * l_1) + b
print("Với width = 5 và length = 10 thì có giá là: ")
print(price_1)
```

```python
#Trường hợp width = 20, length = 20

w_2 = 20 
l_2 = 20
price_2 = (a_1 * w_2 + a_2 * l_2) + b
print("Với width = 20 và length = 20 thì có giá là: ")
print(price_2)
```

```python
#Trường hợp width = 10, length = 40

w_3 = 10
l_3 = 40
price_3 = (a_1 * w_3 + a_2 * l_3) + b
print("Với width = 10 và length = 40 thì có giá là: ")
print(price_3)
```
