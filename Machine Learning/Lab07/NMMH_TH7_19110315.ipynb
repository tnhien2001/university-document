#File data_3.txt

```python
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
def display_data_and_boundary(x,y,pred_fn):  
    # Determine the x1- and x2- limits of the plot
    x1min = min(x[:,0]) - 1
    x1max = max(x[:,0]) + 1
    x2min = min(x[:,1]) - 1
    x2max = max(x[:,1]) + 1
    plt.xlim(x1min,x1max)
    plt.ylim(x2min,x2max)
    # Plot the data points
    k = int(max(y)) + 1
    cols = ['ro', 'k^', 'b*','gx']
    for label in range(k):
        plt.plot(x[(y==label),0], x[(y==label),1], cols[label%4], markersize=8)
        
    # Construct a grid of points at which to evaluate the classifier
    grid_spacing = 0.05
    xx1, xx2 = np.meshgrid(np.arange(x1min, x1max, grid_spacing), np.arange(x2min, x2max, grid_spacing))
    grid = np.c_[xx1.ravel(), xx2.ravel()]
    Z = np.array([pred_fn(pt) for pt in grid])
    # Show the classifier's boundary using a color plot
    Z = Z.reshape(xx1.shape)
    plt.pcolormesh(xx1, xx2, Z, cmap=plt.cm.Pastel1, vmin=0, vmax=k)
    plt.show()
```

```python
#Đọc file data_3.txt
data3 = np.loadtxt("data_3.txt")
# Tạo training set x và label y
x3 = data3[:,0:2]
y3 = data3[:,2]
```

## Train SVM với kernel linear
C thay đổi theo C=[0.1, 1.0, 10.0, 100.0]

###C = 0.1

```python
# Chọn ra 2 features để dễ vẽ hình
features = [0,1]
x3 = x3[:,features]

# Train SVM với kernel linear
clf = SVC(C = 0.1,kernel="linear")
clf.fit(x3,y3)
pred_fn = lambda p: clf.predict(p.reshape(1,-1))
display_data_and_boundary(x3,y3,pred_fn)
```

### C = 1

```python
# Chọn ra 2 features để dễ vẽ hình
features = [0,1]
x3 = x3[:,features]

# Train SVM với kernel linear
clf = SVC(C = 1,kernel="linear")
clf.fit(x3,y3)
pred_fn = lambda p: clf.predict(p.reshape(1,-1))
display_data_and_boundary(x3,y3,pred_fn)
```

### C = 10

```python
# Chọn ra 2 features để dễ vẽ hình
features = [0,1]
x3 = x3[:,features]

# Train SVM với kernel linear
clf = SVC(C = 10,kernel="linear")
clf.fit(x3,y3)
pred_fn = lambda p: clf.predict(p.reshape(1,-1))
display_data_and_boundary(x3,y3,pred_fn)
```

### C = 100.0

```python
# Chọn ra 2 features để dễ vẽ hình
features = [0,1]
x3 = x3[:,features]

# Train SVM với kernel linear
clf = SVC(C = 100,kernel="linear")
clf.fit(x3,y3)
pred_fn = lambda p: clf.predict(p.reshape(1,-1))
display_data_and_boundary(x3,y3,pred_fn)
```

### Nhận xét SVM theo kernel linear:
$⇒$ Qua các mô hình trên với các C > 1, đều không cho thấy sự thay đổi về quá trình phân cụm các dữ liệu, thậm chí là không thay đổi.

$⇒$ Đối với C < 1, cho thấy sự thay đổi rõ rệt về quá trình phân cụm các dữ liệu. 

## Train SVM với kernel rbf

```python
#Sử dụng kernel rbf với gamma lớn, ta có thể thấy các miền quyết định gần như fit vào các điểm dữ liệu

clf = svm.SVC(C = 1, kernel="rbf",gamma=100)
clf.fit(x3,y3)
pred_fn = lambda p: clf.predict(p.reshape(1,-1))
display_data_and_boundary(x3,y3,pred_fn)
```

```python
#Sử dụng kernel rbf, với gamma nhỏ ta thấy miền quyết định rộng hơn

clf = svm.SVC(C = 1, kernel="rbf",gamma=0.5)
clf.fit(x3,y3)
pred_fn = lambda p: clf.predict(p.reshape(1,-1))
display_data_and_boundary(x3,y3,pred_fn)
```

### Nhận xét SVM theo kernel rbf:
- Khi tăng gamma:
 + Lợi: 
   - Giảm thời gian phân cụm.
   - Variance giảm.
 + Hại: 
   - Phân cụm không rõ rệt.
   - Bias tăng.

- Khi giảm gamma:
 + Lợi: 
   - Phân cụm hiệu quả hơn.
   - Bias giảm.
 + Hại: 
   - Thời gian phân cụm khá lâu.
   - Variance tăng.
