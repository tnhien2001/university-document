```python
import pandas as pd
import numpy as np
df = pd.read_csv("House_Price_2.csv") 
df
```

```python
x1 = df['width'].values  
print(x1.shape)  ## trích width thành một mảng numpy
x1 = x1.reshape(-1, 1)                     ## chuyển thành ma trận cột
print(x1.shape)
```

```python
x2 = df['length'].values   ## Trích length 
print(x2.shape)
x2 = x2.reshape(-1,1)
print(x2.shape)
```

```python
y = df['price'].values   ## Trích giá tiền 
print(y.shape)
y = y.reshape(-1,1)
print(y.shape)
```

```python
a, b, c = 1.0, 1.0, 100.0           ## khởi tạo a, b, c

n_iter = 10000           ## số vòng lặp cập nhật a, b, c
lmda = 0.0001         ## learning rate
for ii in range(n_iter):
    error = y - a*x1 - b*x2 - c
    a = a - 2* lmda * np.mean(error*(-x1))    ##cập nhật a và b
    b = b - 2* lmda * np.mean(error*(-x2))
    c = c - 2* lmda * np.mean(error*(-1))
    
    if ii % 1000 == 0:
        loss = np.mean(np.square(y - a*x1 - b*x2 - c))
        print("Loss at iter {}: {}".format(ii, loss))
loss = np.mean(np.square(y - a*x1 - b*x2 - c))
print("Final loss: ", loss)
```

```python
error.shape, x1.shape, x2.shape
```

```python
g_a, g_b, g_c = a, b, c
print(g_a, g_b, g_c)
```

- So sánh với exact solution và Sklearn

```python
## exact solution
import numpy as np
Z = np.concatenate([x1, x2, np.ones((100,1))], axis=-1)   
Z = Z.astype(np.float64)       

## giá trị a và b tìm được
e_a, e_b, e_c = np.matmul(np.matmul(np.linalg.inv(np.matmul(Z.T,Z)), Z.T), y)

print(e_a, e_b, e_c)
```

```python
X = np.concatenate([x1, x2], axis=-1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X, y)                ## train LinearRegression 
sk_a, sk_b, sk_c = lr.coef_[0][0], lr.coef_[0][1], lr.intercept_[0]    ## lấy các tham số học được
print("a = {}, b = {}, c = {}".format(sk_a, sk_b, sk_c))
```

```python
print("GD MSE: ", np.mean(np.square(y - g_a*x1 - g_b*x2 - g_c)))
print("Exact SOl MSE: ", np.mean(np.square(y - e_a*x1 - e_b*x2 - e_c)))
print("SK MSE: ", np.mean(np.square(y - sk_a*x1 - sk_b*x2 - sk_c)))
```
