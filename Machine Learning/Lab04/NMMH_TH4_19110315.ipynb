```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('lienminh.csv')
data.head()
```

```python
## xác suất tiên nghiệm
P_blueWins = 0.5
P_redWins = 0.5

teamWins = data['teamWins'].values
```

# Phân phối Category: firstBlood, heralds, dragons.
- X1: firstBlood.
- X2: heralds.
- X3: dragons.

```python
firstBlood = data['firstBlood'].values
heralds = data['heralds'].values
dragons = data['dragons'].values
```

```python
is_bluefirstBlood = (firstBlood == 'blue').astype(float)
is_blueHeralds = (heralds == 'blue').astype(float)
is_redHeralds = (heralds == 'red').astype(float)
is_blueDragon = (dragons == 'blue').astype(float)
is_redDragon = (dragons == 'red').astype(float)
```

```python
#Tính P(X1|B)
P_blueFirstBlood_blueWins  = np.sum(firstBlood[teamWins == 'blue'] == 'blue') / np.sum(teamWins == 'blue')
P_blueFirstBlood_blueWins = P_blueFirstBlood_blueWins * is_bluefirstBlood + (1-P_blueFirstBlood_blueWins)*(1-is_bluefirstBlood)
P_redFirstBlood_blueWins = 1 - P_blueFirstBlood_blueWins

#Tính P(X2|B)
P_blueHeralds_blueWins  = np.sum(heralds[teamWins == 'blue'] == 'blue') / np.sum(teamWins == 'blue')
P_redHeralds_blueWins  = np.sum(heralds[teamWins == 'blue'] == 'red') / np.sum(teamWins == 'blue')
P_blueHeralds_blueWins = P_blueHeralds_blueWins * is_blueHeralds + P_redHeralds_blueWins * is_redHeralds + (1-P_blueHeralds_blueWins-P_redHeralds_blueWins)*(1-is_blueHeralds-is_redHeralds)
P_redHeralds_blueWins = P_redHeralds_blueWins * is_redHeralds+ P_blueHeralds_blueWins * is_blueHeralds  + (1-P_blueHeralds_blueWins-P_redHeralds_blueWins)*(1-is_blueHeralds-is_redHeralds)
P_noneHeralds_blueWins = 1- P_blueHeralds_blueWins - P_redHeralds_blueWins

#Tính P(X3|B)
P_blueDragons_blueWins  = np.sum(dragons[teamWins == 'blue'] == 'blue') / np.sum(teamWins == 'blue')
P_redDragons_blueWins  = np.sum(dragons[teamWins == 'blue'] == 'red') / np.sum(teamWins == 'blue')
P_blueDragons_blueWins = P_blueDragons_blueWins * is_blueDragon + P_redDragons_blueWins * is_redDragon + (1-P_blueDragons_blueWins-P_redDragons_blueWins)*(1-is_blueDragon-is_redDragon)
P_redDragons_blueWins = P_redDragons_blueWins * is_redDragon+ P_blueDragons_blueWins * is_blueDragon  + (1-P_blueDragons_blueWins-P_redDragons_blueWins)*(1-is_blueDragon-is_redDragon)
P_noneDragons_blueWins = 1- P_blueDragons_blueWins - P_redDragons_blueWins
```

```python
## tính P(X1|R)
P_blueFirstBlood_redWins = np.sum(firstBlood[teamWins == 'red'] == 'blue') / np.sum(teamWins == 'red')
P_blueFirstBlood_redWins = P_blueFirstBlood_redWins * is_bluefirstBlood + (1-P_blueFirstBlood_redWins)*(1-is_bluefirstBlood)

#Tính P(X2|R)
P_blueHeralds_redWins  = np.sum(heralds[teamWins == 'red'] == 'blue') / np.sum(teamWins == 'red')
P_redHeralds_redWins  = np.sum(heralds[teamWins == 'red'] == 'red') / np.sum(teamWins == 'red')
P_blueHeralds_redWins = P_blueHeralds_redWins * is_blueHeralds + P_redHeralds_redWins * is_redHeralds + (1-P_blueHeralds_redWins-P_redHeralds_redWins)*(1-is_blueHeralds-is_redHeralds)
P_redHeralds_redWins = P_redHeralds_redWins * is_redHeralds+ P_blueHeralds_redWins * is_blueHeralds  + (1-P_blueHeralds_redWins-P_redHeralds_redWins)*(1-is_blueHeralds-is_redHeralds)
P_noneHeralds_redWins = 1- P_blueHeralds_redWins - P_redHeralds_redWins

#Tính P(X3|B)
P_blueDragons_redWins  = np.sum(dragons[teamWins == 'red'] == 'blue') / np.sum(teamWins == 'red')
P_redDragons_redWins  = np.sum(dragons[teamWins == 'red'] == 'red') / np.sum(teamWins == 'red')
P_blueDragons_redWins = P_blueDragons_redWins * is_blueDragon + P_redDragons_redWins * is_redDragon + (1-P_blueDragons_redWins-P_redDragons_redWins)*(1-is_blueDragon-is_redDragon)
P_redDragons_redWins = P_redDragons_redWins * is_redDragon+ P_blueDragons_redWins * is_blueDragon  + (1-P_blueDragons_redWins-P_redDragons_redWins)*(1-is_blueDragon-is_redDragon)
P_noneDragons_redWins = 1- P_blueDragons_redWins - P_redDragons_redWins
```

```python
## P(B|X)
P_blueWins_X = (P_blueFirstBlood_blueWins *P_blueHeralds_blueWins*P_blueDragons_blueWins* P_blueWins ) / ((P_blueFirstBlood_blueWins*P_blueHeralds_blueWins*P_blueDragons_blueWins * P_blueWins) + 
              (P_blueFirstBlood_redWins* P_blueHeralds_redWins* P_blueDragons_redWins* P_redWins))
acc = np.mean((P_blueWins_X > 0.5) == (teamWins == 'blue'))
print(acc)
```

```python
X = data.iloc[:,3:6].values

## Dùng sklearn
from sklearn.naive_bayes import CategoricalNB
blueX_float = (X.reshape(-1,3) == 'blue').astype(float)
nb = CategoricalNB(alpha=1.0e-10).fit(blueX_float, teamWins)
np.mean(teamWins == nb.predict(blueX_float))
```

# Phân phối normal: 
- X1: killsDiff.
- X2: minionsKilledDiff.
- X3: wardPlacedDiff.

```python
killsDiff = data['killsDiff'].values
minionsKilledDiff = data['minionsKilledDiff'].values
wardPlaceDiff = data['wardPlacedDiff'].values
```

```python
## tính tham số
killsDiff_mean_blueWins = np.mean(killsDiff[teamWins == 'blue'])
killsDiff_std_blueWins = np.std(killsDiff[teamWins == 'blue'])

killsDiff_mean_redWins = np.mean(killsDiff[teamWins == 'red'])
killsDiff_std_redWins = np.std(killsDiff[teamWins == 'red'])

minionsKilledDiff_mean_blueWins = np.mean(minionsKilledDiff[teamWins == 'blue'])
minionsKilledDiff_std_blueWins = np.std(minionsKilledDiff[teamWins == 'blue'])

minionsKilledDiff_mean_redWins = np.mean(minionsKilledDiff[teamWins == 'red'])
minionsKilledDiff_std_redWins = np.std(minionsKilledDiff[teamWins == 'red'])

wardPlacedDiff_mean_blueWins = np.mean(wardPlacedDiff[teamWins == 'blue'])
wardPlacedDiff_std_blueWins = np.std(wardPlacedDiff[teamWins == 'blue'])

wardPlacedDiff_mean_redWins = np.mean(wardPlacedDiff[teamWins == 'red'])
wardPlacedDiff_std_redWins = np.std(wardPlacedDiff[teamWins == 'red'])
```

```python
def normal(x, mean, std):
    return 1/(np.sqrt(2*np.pi)*std)*np.exp(-np.square(x-mean)/(2*np.square(std)))
```

```python
## P(X1|B) và P(X1|R)
P_minionsKilledDiff_blueWins = normal(minionsKilledDiff, minionsKilledDiff_mean_blueWins, minionsKilledDiff_std_blueWins)
P_minionsKilledDiff_redWins = normal(minionsKilledDiff, minionsKilledDiff_mean_redWins, minionsKilledDiff_std_redWins)
## P(X2|B) và P(X2|R)
P_killsDiff_blueWins = normal(killsDiff, killsDiff_mean_blueWins, killsDiff_std_blueWins)
P_killsDiff_redWins = normal(killsDiff, killsDiff_mean_redWins, killsDiff_std_redWins)
## P(X3|B) và P(X3|R)
P_wardPlacedDiff_blueWins = normal(wardPlacedDiff, wardPlacedDiff_mean_blueWins, wardPlacedDiff_std_blueWins)
P_wardPlacedDiff_redWins = normal(wardPlacedDiff, wardPlacedDiff_mean_redWins, wardPlacedDiff_std_redWins)
```

```python
## P(B|X)
P_blueWins_X = (P_minionsKilledDiff_blueWins*P_killsDiff_blueWins* P_wardPlacedDiff_blueWins* P_blueWins) / ((P_minionsKilledDiff_blueWins*P_killsDiff_blueWins* P_wardPlacedDiff_blueWins * P_blueWins) + (P_minionsKilledDiff_redWins*P_killsDiff_redWins*P_wardPlacedDiff_redWins * P_redWins))
## accuracy
acc = np.mean((P_blueWins_X > 0.5) == (teamWins == 'blue'))
acc
```
