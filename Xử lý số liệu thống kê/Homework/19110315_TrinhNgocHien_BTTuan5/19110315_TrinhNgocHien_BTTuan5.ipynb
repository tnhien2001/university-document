### Indexing, selection, and filtering

### Function application and mapping

```python
frame = DataFrame(np.random.randn(4,3), columns=list('bde'),
                 index=['Utah','Ohio','Texas','Oregon'])
print(frame)
print("--------------")
abs = np.abs(frame) #Trị tuyệt đối của frame
print(abs) 
print("--------------")

print("--------------")
f = lambda x: x.max() - x.min()
lambda1 = frame.apply(f)
print(lambda1)
print("--------------")
lambda2 = frame.apply(f,axis=1)
print(lambda2)
print("--------------")

print("--------------")
def f(x):
    return Series([x.min(),x.max()], index=['min','max'])
t = frame.apply(f)
print(t)
print("--------------")

print("--------------")
format = lambda x: '%.2f' % x
lambda3 = frame.applymap(format)
print(lambda3)
print("--------------")
t1 = frame['e'].map(format)
print(t1)
```

# 3.5 Chỉ số phân cấp - Hierarchical Indexing

```python
data = Series(np.random.randn(10),
             index=[['a','a','a','b','b','b','c','c','d','d'],
                   [1,2,3,1,2,3,1,2,2,3]])
print(data)
print("--------------")
index = data.index
print(index)
```

```python
print(data['b'])
print("--------------")
print(data['b':'c'])
print("--------------")
print(data.loc[['b','d']])
print("--------------")
data[:,2]
```

```python
#Sắp xếp Series có chỉ mục phân cấp thành DataFrame và ngược lại

data1=data.unstack() #Xếp về dạng ngang
print(data1)
print("--------------")
data2=data.unstack().stack() #Xếp về dạng dọc
print(data2)
```
