# 5.5 Vẽ một số hình ảnh trong Python

```python
import numpy as np
canvas = np.zeros((300, 300, 3), dtype = "uint8")
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
cv2_imshow(canvas)
cv2.waitKey(0)
```

```python
red = (0, 0, 255)
cv2.line(canvas, (300,0), (0,300), red, 3)
cv2_imshow(canvas)
cv2.waitKey(0)
```

```python
# Vẽ HCN:
cv2.rectangle(canvas, (10,10), (60,60), green)
cv2_imshow(canvas)
cv2.waitKey(0)
```

```python
cv2.rectangle(canvas, (50,200), (200,225), red,5)
cv2_imshow(canvas)
cv2.waitKey(0)
```

```python
blue = (255, 0, 0)
cv2.rectangle(canvas, (200,50), (225,125), blue, -1)
cv2_imshow(canvas)
cv2.waitKey(0)
```

```python
# Vẽ đường tròn
canvas = np.zeros ((300, 300, 3) ,dtype = "uint8" )
(centerX , centerY) = (canvas.shape[1] // 2 ,canvas.shape[0] // 2)
white = (255 , 255 , 255 )

for r in range(0,175,25):
  cv2.circle(canvas, (centerX, centerY), r, white)
  
cv2_imshow(canvas)
cv2.waitKey(0)
```

```python
# Bài tập:
import matplotlib.pyplot as plt
canvas = np.ones([300, 300, 3])

```

```python
#Tam giác vuông 1:
blue = (0, 0, 255)
cv2.line(canvas, (20,20), (20,40), blue)
cv2.line(canvas, (20,20), (50,40), blue)
cv2.line(canvas, (20,40), (50,40), blue)
plt.imshow(canvas)


#Tam giác đều 2:
green = (0, 255, 0)
cv2.line(canvas, (80, 20), (67,33), green)
cv2.line(canvas, (80,20), (93,33), green)
cv2.line(canvas, (67,33), (93,33), green)
plt.imshow(canvas)

#Hình vuông
red = (255, 0, 0)
cv2.line(canvas, (20, 60), (20,100), red)
cv2.line(canvas, (20, 60), (60,60), red)
cv2.line(canvas, (60,60), (60,100), red)
cv2.line(canvas, (20,100), (60,100), red)
plt.imshow(canvas)

#Hình bình hành:
black = (0,0,0)
cv2.line(canvas, (80, 60), (100,60), black)
cv2.line(canvas, (100, 60), (95,70), black)
cv2.line(canvas, (95, 70), (75,70), black)
cv2.line(canvas, (75, 70), (80,60), black)
plt.imshow(canvas)
```
