```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = 'data_w1.csv'
path = pd.read_csv(data)
df = pd.DataFrame(path)
df
```

```python
df.describe()
```

```python

x = df.iloc[:,:30].values
x.shape
```

```python
x
```

```python
x_scaled = x - np.mean(x,axis = 0)
x_scaled
```

```python
np.mean(x_scaled, axis = 0)
```

```python
x.shape
```

```python
features = x_scaled.T
cov_matrix = np.cov(features)
cov_matrix
```

```python
values, vectors = np.linalg.eig(cov_matrix)
```

```python
print(values)
vectors
```

```python
vectors[:,0].dot(vectors[:,29])
```

```python
u,s,vh = np.linalg.svd(x_scaled)
vh.T
```

```python
smat = np.zeros(x_scaled.shape, dtype = complex)
smat[:30, :30] = np.diag(s)
(1/x_scaled.shape[0])*np.matmul(smat, smat.T)[:30,:30]
```

```python
explained_variances = []
for i in range(len(values)):
  explained_variances.append(values[i] / np.sum(values))

print(np.sum(explained_variances), '\n', explained_variances)
```

```python
projected_1 = x_scaled.dot(vectors[:,0])
projected_2 = x_scaled.dot(vectors[:,1])
res = pd.DataFrame(projected_1, columns=['PC1'])
res['PC2'] = projected_2

res.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (20, 10))
sns.scatterplot(res['PC1'], [0] * len(res), s = 200)
```
