```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data_w1.csv")
data.head()
```

```python
data.describe()
```

```python
df_num = data.to_numpy()
```

```python
data_R=[]
for i in range(len(df_num)-1):
    temp = (df_num[i+1] - df_num[i])/df_num[i]
    data_R.append(temp)

print(type(data_R))
```

```python
np_data_R = np.asarray(data_R) 
print(type(np_data_R))
print(np_data_R.shape)
```

```python
#Tính kỳ vọng
kyvong = np.mean(np_data_R,axis = 0)
kyvong
```

```python
X_scaled = np_data_R - kyvong
X_scaled
```

```python
cov_matrix = np.cov(X_scaled.T)
cov_matrix
```

```python
value , vector = np.linalg.eig(cov_matrix)
```

```python
def pca_alg(data):
  #tính vector trung bình dữ liệu
  kyvong = np.mean(data,axis = 0)

  #Trừ mỗi điểm dữ liệu cho vector trung bình
  X_hat = (data - kyvong)

  #Tính ma trận hiệp phương sai
  features = X_hat.T
  cov_matrix = np.cov(features)

  #Tính các cặp trị riêng, vector riêng và sắp xếp theo thứ tự giảm dần tri riêng
  eig_values, eig_vectors = np.linalg.eig(cov_matrix)
  #Sắp xếp giấ trị theo chiều giảm dần
  sort_index = np.argsort(eig_values)[::-1]
  new_eig_values = eig_values[sort_index]
  new_eig_vectors = eig_vectors[:,sort_index]
  
  return new_eig_values, new_eig_vectors
```

```python
new_eig_values, new_eig_vectors = pca_alg(np_data_R)
```
