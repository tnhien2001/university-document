```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data_w1.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
df_num = data.to_numpy()
```

```python
data_R=[]
for i in range(len(df_num)-1):
    temp = (df_num[i+1] - df_num[i])/df_num[i]
    data_R.append(temp)

print(type(data_R))
```

```python
np_data_R = np.asarray(data_R) 
print(type(np_data_R))
print(np_data_R.shape)
```

```python
from sklearn.preprocessing import StandardScaler
data = np_data_R
scaler = StandardScaler()
scaler.fit(data)

X_scaled = scaler.transform(data)
X_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(X_scaled)
# print(pca.explained_variance_ratio_)
```
