```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.gridspec as gridspec
```

```python
data = sns.load_dataset('iris')
display(data.head())
print(data.shape)
```

# LDA (Linear Discriminant Analysis)


1) Tính ma trận phân tán bên trong lớp:

$S_W = \sum_{c} S_c$

$S_c = \sum_{i\in{c}}(x_i - \overline{x}_c) \,\,.\,(\overline{x}_c - \overline{x})^T$

2) Tính khoảng giữa ma trận phân tán lớp:

$S_B = \sum_{c} n_c \, . \, (\overline{x}_c - \overline{x}) \, . \, (\overline{x}_c - \overline{x})^T$





```python
x = data.iloc[:,:4].values
```

```python
y = []
for i in data['species']:
  if i =='setosa':
    y.append(0)
  if i == 'versicolor':
    y.append(1)
  if i == 'virginica':
    y.append(2)

y = np.array(y)
labels = np.unique(y)
```

```python
n_features = x.shape[1]
n_features
```

```python
mean_overall = np.mean(x, axis=0)
mean_overall
```

### Khởi tạo $S_W$ và $S_B$ và tiến hành tính:

```python
S_W = np.zeros((n_features, n_features))
S_B = np.zeros((n_features, n_features))


for c in labels:
  x_c = x[y==c]
  mean_c = np.mean(x_c, axis=0)
  S_W += (x_c - mean_c).T.dot(x_c - mean_c)

  n_c = x_c.shape[0]
  mean_diff = (mean_c - mean_overall).reshape(n_features, 1)
  S_B += n_c * (mean_diff).dot(mean_diff.T)


```

### Tính trị riêng và vector riêng sau đó sắp xếp theo thứ tự:

Với Trị riêng và Vector riêng được tính theo $S_W^{-1} S_B$

```python
A = np.linalg.inv(S_W).dot(S_B)
eigenvalues, eigenvectors = np.linalg.eig(A)
eigenvectors = eigenvectors.T
#Sắp xếp vecto riêng và giá trị riêng theo thứ tự từ bé đến lớn
idxs = np.argsort(abs(eigenvalues))[::-1]
eigenvalues = eigenvalues[idxs]
eigenvectors = eigenvectors[idxs]
print("Eigenvalues: ",eigenvalues,"\n","\n","Eigenvectors: ",eigenvectors)
```

```python
explained_variances = []
for i in range(len(eigenvalues)):
    explained_variances.append(eigenvalues[i] / np.sum(eigenvalues))
 
print("Sum of Explained Variances: ",np.sum(explained_variances), '\nExplained Variances:', explained_variances)
```

```python
print("Vecto riêng ứng với PC1", eigenvectors[:,0])
print("Vecto riêng ứng với PC2", eigenvectors[:,1])
```

# PCA (Principal Components Analysis)


```python
X = data.iloc[:,:4].values
```

```python
def pca_alg(data):

  #tính vector trung bình dữ liệu
  
  kyvong = np.mean(data,axis = 0)

  #Trừ mỗi điểm dữ liệu cho vector trung bình
  X_std = (data - kyvong)/np.std(data,axis = 0)

  #Tính ma trận hiệp phương sai
  features = X_std.T
  cov_matrix = np.cov(features)

  #Tính các cặp trị riêng, vector riêng và sắp xếp theo thứ tự giảm dần tri riêng
  values, vectors = np.linalg.eig(cov_matrix)
  eig_values = values[np.argsort(values)[::-1]]
  eig_vectors = vectors[:,np.argsort(values)[::-1]]

  return eig_values, eig_vectors
```

```python
eig_values, eig_vectors = pca_alg(X)
print("Eig_values: ",eig_values)
print("Eig_vectors: ",eig_vectors)
```

```python
explained_variances = []
for i in range(len(eig_values)):
    explained_variances.append(eig_values[i] / np.sum(eig_values))
 
print("Sum of Explained Variances: ",np.sum(explained_variances), '\nExplained Variances:', explained_variances)
```

```python
print("Vecto riêng ứng với PC1", eig_vectors[:,0])
print("Vecto riêng ứng với PC2", eig_vectors[:,1])
```

### So sánh 2 thuật toán LDA và PCA:
#### - Thuật toán LDA:
- Là thuật toán thuộc dạng Supervised Learning.
- Quan tâm đến các trục tối đa hóa sự tách biệt giữa nhiều lớp

#### - Thuật toán PCA:
- Là thuật toán thuộc dạng Unsupervised Learning.
- Tìm các trục thành phần tối đa hóa phương sai của dữ liệu.
