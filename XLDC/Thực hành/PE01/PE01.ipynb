## PCA

- Kiểm tra và tính toán các giá trị thống kê của dữ liệu.

```python
df.describe()
```

- Trích các dữ liệu FrontLeft, FrontRight, BackLeft, BackRight từ dữ liệu lớn.

```python
#Xử lý dữ liệu từ bảng:
x = df.iloc[:,1:].values
print(x.shape)
print(type(x))
```

- Thiết lập hàm PCA để tính vectơ riêng và giá trị riêng của dữ liệu.

```python
def pca_alg(data):

  #tính vector trung bình dữ liệu
  
  kyvong = np.mean(data,axis = 0)

  #Trừ mỗi điểm dữ liệu cho vector trung bình
  X_std = (data - kyvong)/np.std(data,axis = 0)

  #Tính ma trận hiệp phương sai
  features = X_std.T
  cov_matrix = np.cov(features)

  #Tính các cặp trị riêng, vector riêng và sắp xếp theo thứ tự giảm dần tri riêng
  values, vectors = np.linalg.eig(cov_matrix)
  eig_values = values[np.argsort(values)[::-1]]
  eig_vectors = vectors[:,np.argsort(values)[::-1]]

  return eig_values, eig_vectors
```

```python
eig_values, eig_vectors = pca_alg(x)
print("Eig_values: ",eig_values)
print("Eig_vectors: ",eig_vectors)
```

```python
explained_variances = []
for i in range(len(eig_values)):
    explained_variances.append(eig_values[i] / np.sum(eig_values))
 
print("Sum of Explained Variances: ",np.sum(explained_variances), '\nExplained Variances:', explained_variances)
```

```python
print("Vecto riêng ứng với PC1", eig_vectors[:,0])
print("Vecto riêng ứng với PC2", eig_vectors[:,1])
```

- Giảm số chiều của dữ liệu.

```python
projected_1 = x.dot(eig_vectors[:,0])
projected_2 = x.dot(eig_vectors[:,1])
res = pd.DataFrame({"PC1": projected_1,
                   "PC2": projected_2})
res
```

- Vẽ hình:  PC1 và PC2



```python
#PC1:

plt.figure(figsize =(20,10))
plt.plot(res['PC1'])
```

```python
#PC2:

plt.figure(figsize =(20,10))
plt.plot(res['PC2'])
```

# Code với sklearn

```python
from sklearn.decomposition import PCA

# n_components = 2
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(x)
```

```python
X_pca_2 = -1*(X_pca)
temp = pd.DataFrame(X_pca_2, columns=['PC1','PC2'])
temp['PC2'] = -1*temp['PC2']
temp
```

```python
plt.figure(figsize =(20,10))
plt.plot(temp)
```

# Kết luận:
- Đối với n_component = 1 (PC1, màu xanh), thì hình vẽ được trực quan hóa dữ liệu từ n_component = 1 (PC1) (từ thông thường cho đến sklearn) đều cho kết quả gần tương đồng so FrontLeft, FrontRight, BackLeft, BackRight. 
- Đối với n_component = 2 (PC2, màu cam), thì hình vẽ được trực quan hóa dữ liệu khác rất nhiều so với FrontLeft, FrontRight, BackLeft, BackRight.

=> Chọn n_component = 1 vì sẽ giữ được nhiều thông tin nhất có thể.
