```python
# PT 1
%%writefile pt1.py
def ptbac1(a, b):
    '''
    Ham giai phuong trinh bac 1

    parametes:
        a : 

        b :

    return: (so_nghiem, list_nghiem)
        so_nghiem:
            -1 : 'Vo so nghiem
            0 : 'Vo ngiem'
            1 : '1 nghiem'
        list_nghiem:
            []
            [ 1 ] : nghiem la 1
    '''
    # if a == 0:
    #     if b == 0:
    #         return -1, None

    #     else:
    #         return 0, None
    
    # else:
    #     return 1, -b/a

    # Optim
    if not a:
        if not b:
            return -1, []

        return 0, []
    
    return 1, [-b/a]
```

```python
# ax2 + bx +c = 0 
%%writefile pt2.py
from math import sqrt
from pt1 import ptbac1

def ptbac2(a, b, c):
    '''
    Giải phương trình bậc 2: ax2 + bx + c = 0

    parameters:

        ......
    
    return:
        (so_nghiem, list_nghiem)
        so_nghiem:
            -1 : 'Vo so nghiem
            0 : 'Vo ngiem'
            1 : '1 nghiem'
            2 : '2 nghiem'
        list_nghiem:
            [] : 
            [ , ] :  
    '''

    # if not a:
    #     return ptbac1(b,c) 

    # delta = b ** b - 4 * a * c

    # # delta = 0, 1 nghiem, 
    # if not delta:
    #     return 1, [-b / (2 * a)]
    
    # # delta < 0, 0 nghiem, 
    # if delta < 0:
    #     return 0, None
    
    # # delta > 0, 2 nghiem
    # return 2, [(-b + sqrt(delta)) / (2 * a), (-b - sqrt(delta)) / (2 * a)]

    # # optim  -----------------------------
    # # chỉ dùng 1 return

    if not a:
        so_nghiem, list_nghiem =  ptbac1(b,c) 

    else:
        delta = b * b - 4 * a * c

        # delta = 0, 1 nghiem, 
        if not delta:
            so_nghiem, list_nghiem = 1, [-b / (2 * a)]
        
        # delta < 0, 0 nghiem, 
        elif delta < 0:
            so_nghiem, list_nghiem = 0, None
        
        # delta > 0, 2 nghiem
        else:
            so_nghiem, list_nghiem =  2, [(-b + sqrt(delta)) / (2 * a), (-b - sqrt(delta)) / (2 * a)]
    
    return so_nghiem, list_nghiem
```

```python
# 
%%writefile run.py
import sys
from pt2 import *

try:
    a, b, c = float(sys.argv[1]), float(sys.argv[2]), float(sys.argv[3])

    so_nghiem, list_nghiem = ptbac2(a,b,c)

    # print('So luong nghiem : ' , so_nghiem)
    
    if so_nghiem == -1:
        print("Vo so nghiem")
    else:
        print(f'So luong nghiem : {so_nghiem}')

    if so_nghiem > 0:
        for nghiem in list_nghiem:
            print(f'Nghiem : {nghiem}')
except IndexError:
    print('Loi')

finally: 
    print('Ket thuc')
```
