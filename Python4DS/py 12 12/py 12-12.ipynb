```python
from faker import Faker
import pandas as pd

f = Faker()

df_sample = pd.DataFrame()
df_sample['name'] = [f.name() for itr in range(50)]
df_sample
```

```python
import sqlalchemy
import pandas as pd

USERNAME = 'postgres'
PASSWORD = '12345678'
HOST = 'localhost'
PORT = 5432
DBNAME = 'dblocal'

sql_connect = f'postgresql+psycopg2://{USERNAME}:%s@{HOST}:{PORT}/{DBNAME}' %PASSWORD


conn = sqlalchemy.create_engine(sql_connect)
```

```python
df_sample.to_sql('dblocal', con=conn, if_exists='append', index = False, method='multi')
```

```python
df = pd.read_sql('Select * From dblocal', conn)
df
```
