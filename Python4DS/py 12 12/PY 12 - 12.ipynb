```python
df = pd.read_sql('Select * From dblocal WHERE year > 10', conn)
```

```python
df['address'].iloc[16]
```

```python
# --> Clean data
# Parsing đường, bang, code, ...
import re
parser = df['address'].apply(lambda itr: [i for i in re.findall('(.*)\n(.*)', itr)[0]])
```

```python
street = []
district = []
for add in df['address']:
#     re.findall('(.*)\n(.*)', add)
    parser = add.split('\n')
    street.append(parser[0])
    district.append(parser[1])
    
```

```python
df['street'] = street
df['district'] = district
```

```python
df_street = pd.DataFrame()
df_street['street'] = df['street']
```

```python
dict_street = dict()
for index, st in enumerate(df_street['street'].unique()):
    dict_street[st] = index
```

```python
dict_street
```

```python
df_street['id'] = df_street['street'].apply(lambda x: dict_street[x])
df['id_street'] = df['street'].apply(lambda x: dict_street[x])
```

```python
del df['street']
```

```python
df.to_sql('dblocal__replace_street', con=conn, if_exists='append', index = False)
```

```python
df_street.to_sql('dblocal__street', con=conn, if_exists='append', index = False)
```

```python
# Lây xuống
sql_query = '''
SELECT * FROM dblocal__replace_street, dblocal__street
WHERE dblocal__replace_street.id_street = dblocal__street.id
'''

df_new = pd.read_sql(sql_query, conn)
df_new
```
