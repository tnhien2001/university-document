# Cancel with break

```python
count = 1
while count < 10:
    print(count)

    if count == 6:
        break
        
    count += 1 
```

```python
itr_1 = 0
while itr_1 < 3:
    itr_2 = 0
    while itr_2 < 3:
        print(itr_1, itr_2)

        if itr_2 == 1 and itr_1 == 1:
            break
        
        itr_2 += 1

    itr_1 += 1
```

# Iterate Multiple Sequences with zip()

```python
# lấy index của 1 list theo thứ tự tăng dần hoặc giảm dần.
```

```python
days = ['Monday', 'Tuesday', 'Wednesday']
fruits = ['banana', 'orange', 'peach']
drinks = ['coffee', 'tea', 'beer']
desserts = ['tiramisu', 'ice cream', 'pie', 'pudding']

print('moethos 1')
for day, fruit, drink, dessert in zip(days, fruits, drinks, desserts):
    print(day, ": drink", drink, "- eat", fruit, "- enjoy", dessert)

print('methos 2')
for itr_zip in zip(days, fruits, drinks, desserts):
    print(itr_zip)

print('moethos 3') #
for day, fruit, drink, dessert in zip(days, fruits, drinks, desserts):
    print(f'{day}: drink {drink} - eat {fruit} - enjoy {dessert}')
```

### Gather Positional Arguments with *

```python
def print_args(*args):
    print('Positional argument tuple:', args)

print_args(3, 2, 1, 'wait!', 'uh...', '1', '2')
```

```python
def print_args(*args):
    print('Phần tử vị trí 2:', args[2])

print_args(3, 2, 1, 'wait!', 'uh...', '1', '2')
```

### Gather Keyword Arguments with **

```python
def print_kwargs(**kwargs):
    print('Keyword arguments:', kwargs)
```

```python
print_kwargs(wine='merlot', entree='mutton', dessert='macaroon')
```

```python
# print value of wine
def print_kwargs(**kwargs):
    'Something'
    print('Keyword arguments:', kwargs['wine'])

print_kwargs(wine='merlot', entree='mutton', dessert='macaroon')
```

```python
# error "không có biên wine"
print_kwargs(entree='mutton', dessert='macaroon')
```
