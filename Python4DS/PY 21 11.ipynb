```python
# https://shopee.vn/api/v4/microsite/campaign_site_page?url=uu-dai-unilever&platform=pc&_mod=microsite

import requests
url = 'https://shopee.vn/api/v4/microsite/campaign_site_page?url=uu-dai-unilever&platform=pc&_mod=microsite'
rp = requests.get(url)

rp.content
```
