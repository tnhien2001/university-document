```python
##LIST:
from collections import defaultdict

graph = defaultdict(list)

graph['[3,3,1]'] = ['[2,2,0]', '[3,1,0]']

graph['[2,2,0]'] = ['[3,2,1]']
graph['[3,1,0]'] = ['[3,2,1]']


graph['[3,2,1]'] = ['[2,2,0]', '[3,1,0]', '[3,0,0]']

graph['[3,0,0]'] = ['[3,1,1]']

graph['[3,1,1]'] = ['[1,1,0]']

graph['[1,1,0]'] = ['[2,2,1]']

graph['[2,2,1]'] = ['[0,2,0]']

graph['[0,2,0]'] = ['[0,3,1]']

graph['[0,3,1]'] = ['[0,1,0]']

graph['[0,1,0]'] = ['[1,1,1]', '[0,2,1]']

graph['[0,2,1]'] = ['[0,0,0]']
graph['[1,1,1]'] = ['[0,0,0]']
```

```python
class Node:
    def __init__(self, name, par = None):
        self.name = name
        self.par = par
        
    def display(self):
        print(self.name)

def equal(O, G):
    return O.name == G.name

def checkInArray(tmp, open):
    for x in open:
        if equal(x, tmp):
            return True
    return False

def path(O):
    print(O.name)
    if O.par != None:
        path(O.par)
    else:
        return
```

```python
def bfs(S = Node('[3,2,1]'), G = Node('[0,0,0]')):
    open = []
    closed = []

    #Thêm nút vào 
    open.append(S)
    while True:
        if len(open) == 0: #Nếu ngăn rỗng
            print("Tim kiem that bai")
            return
        O = open.pop(0)
        closed.append(O)
        
        print("Duyet: ",O.name)
        if equal(O, G) == True:
            print("Tim kiem thanh cong")
            path(O)
            return 
        
        for x in graph[O.name]:
            tmp = Node(x)
            tmp.par = O
            
            ok1 = checkInArray(tmp, open)
            ok2 = checkInArray(tmp, closed)
        #Nếu phần tử tạm không nằm trong open và close thì đưa vào cuối open
        if not ok1 and not ok2:
            open.append(tmp)
```

```python
bfs(Node('[3,1,1]'), Node('[0,0,0]'))
```
