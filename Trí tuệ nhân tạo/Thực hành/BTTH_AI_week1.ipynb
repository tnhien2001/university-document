# Breadth-first search




```python
class Node:
    def __init__(self, name, par = None):
        self.name = name
        self.par = par
        
    def display(self):
        print(self.name)
```

```python
from collections import defaultdict

graph = defaultdict(list)
graph['V1'] = ['V2','V4','V3']
graph['V2'] = ['V7']
graph['V4'] = ['V5']
graph['V3'] = ['V6','V8']
graph['V7'] = ['V3']
graph['V5'] = ['V14']
graph['V6'] = ['V12']
graph['V8'] = ['V10','V9']
graph['V14'] = ['V11']
graph['V12'] = ['V13']
graph['V10'] = []
graph['V9'] = ['V11']
graph['V11'] = ['V15']
graph['V13'] = ['V14']
graph['V15'] = ['V16']
graph['V16'] = ['V17']
graph['V17'] = ['V18']
graph['V18'] = []
```

```python
def equal(O, G):
    return O.name == G.name

def checkInArray(tmp, open):
    for x in open:
        if equal(x, tmp):
            return True
    return False

def path(O):
    print(O.name)
    if O.par != None:
        path(O.par)
    else:
        return
```

```python
def bfs(S = Node('V1'), G = Node('V18')):
    open = []
    closed = []

    #Thêm nút vào 
    open.append(S)
    while True:
        if len(open) == 0: #Nếu ngăn rỗng
            print("Tim kiem that bai")
            return
        O = open.pop(0)
        closed.append(O)
        
        print("Duyet: ",O.name)
        if equal(O, G) == True:
            print("Tim kiem thanh cong")
            path(O)
            return 
        
        for x in graph[O.name]:
            tmp = Node(x)
            tmp.par = O
            
            ok1 = checkInArray(tmp, open)
            ok2 = checkInArray(tmp, closed)
        #Nếu phần tử tạm không nằm trong open và close thì đưa vào cuối open
        if not ok1 and not ok2:
            open.append(tmp)
```

```python
bfs(Node('V1'),Node('V18'))
```

# Depth-first search

```python
def dfs(S = Node('A'), G = Node('M')):
  open = []
  closed = []
  open.append(S)

  while True:
    if len(open) == 0:
      print("Tim kiem that bai")
      return 

    O = open.pop(0)
    closed.append(O)

    print("Duyet: ",O.name)
    if equal(O,G):
      print("Tim thay")
      path(O)
      return
    
    pos = 0
    for x in graph[O.name]:
      tmp = Node(x)
      tmp.par = O

    ok1 = checkInArray(tmp,open)
    ok2 = checkInArray(tmp,closed)
    if not ok1 and not ok2:
      open.insert(pos,tmp)
      pos += 1
```

```python
dfs(Node('V1'),Node('V18'))
```

# Uniform Cost Search

```python
class Node:
    def __init__(self, name, par = None, g = 0):
        self.name = name
        self.par = par
        self.g = g
        
    def display(self):
        print(self.name, self.g)

#less than
    def __lt__(self,other):
        if other == None:
          return False
        return self.g < other.g

    def __eq__(self,other):
      if other == None:
        return False
      return self.name == other.name
```

```python
def equal(O,G):
  if O.name == G.name:
    return True
  return False

def checkInPriority(tmp,c):
  if tmp == None:
    return False
  return (tmp in c.queue)

def getPath(O):
  print(O.name)
  if O.par != None:
    getPath(O.par)
  else:
    return
    
```

```python
def ucs(S = Node('A'), G = Node('N')):
  open = PriorityQueue()
  close = PriorityQueue()
  open.put(S)
  while True:
    if open.empty():
      print("Tim kiem that bai")
      return

    O = open.get()
    close.put(O)
    print('duyet: ', O.name, O.g)
    if equal(O,G) == True:
      print('Tim kiem thanh cong')
      getPath(O)
      print('distance: ', O.g)
      return
    
    i = 0
    while i<len(graph[O.name]):
      name = graph[O.name][i]
      g = O.g + graph[O.name][i+1]
      tmp = Node(name = name, g = g)
      tmp.par = O
      
      ok1 =  checkInPriority(tmp,open)
      ok2 =  checkInPriority(tmp,close)

      if not ok1 and not ok2:
        open.put(tmp)
      
      i += 2 
```

```python
ucs(Node('V1'),Node('V18'))
```
